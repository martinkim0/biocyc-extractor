```python
f = open('inputs/ecoliW/reactions.dat', encoding='ISO-8859-1')

id, left, right, direction = (None, [], [], None)
while True:
    line = f.readline()
    if not line:
        break
    if line.startswith('//'):
        print(id, left, right, direction)
        id, left, right, direction = (None, [], [], None)
        continue

    entries = line.split(' - ')
    entry_type = entries[0].strip()
    if entry_type == 'UNIQUE-ID':
        id = entries[1].strip()
    elif entry_type == 'LEFT':
        left.append(entries[1].strip())
    elif entry_type == 'RIGHT':
        right.append(entries[1].strip())
    elif entry_type == 'REACTION-DIRECTION':
        dir = entries[1].strip()
        if dir == 'REVERSIBLE':
            direction = 0
        elif dir.endswith('LEFT-TO-RIGHT'):
            direction = 1
        elif dir.endswith('RIGHT-TO-LEFT'):
            direction = 2
        else:
            direction = -1
```
